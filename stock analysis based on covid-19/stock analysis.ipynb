```python
start = datetime.datetime(2020, 1, 1)
end = datetime.datetime(2020, 8, 17)
```

```python
DAL=data.DataReader("DAL",'yahoo',start,end)
#Delta Airlines Holdings, Inc.
UAL=data.DataReader("UAL",'yahoo',start,end)
#United Airlines Holdings, Inc. 
AAL=data.DataReader("AAL",'yahoo',start,end)
#American Airlines Group Inc.
LUV=data.DataReader("LUV",'yahoo',start,end)

tickers1 = ['DAL', 'UAL', 'AAL', 'LUV']
Airlines = pd.concat([DAL, UAL, AAL, LUV],axis=1,keys=tickers1)
Airlines.columns.names=['Stock name','Stock Info']
Airlines.describe()

Airlines
Airlines.xs(key='Close',axis=1,level='Stock Info').max()

```

```python
MSFT=data.DataReader("MSFT",'yahoo',start,end)
#Microsoft Corp
IBM=data.DataReader("IBM",'yahoo',start,end)
#International Business Machines
FB=data.DataReader("FB",'yahoo',start,end)
#Facebook Inc
AAPL=data.DataReader("AAPL",'yahoo',start,end)
#Apple INC

tickers2 = ['MSFT', 'IBM', 'FB', 'AAPL']
Technology = pd.concat([MSFT, IBM, FB, AAPL],axis=1,keys=tickers2)
Technology.columns.names=['Stock name','Stock Info']
Technology.describe()

Technology
```

```python
JPM=data.DataReader("JPM",'yahoo',start,end)
#JPMorgan Chase & Co.
GS=data.DataReader("GS",'yahoo',start,end)
#The Goldman Sachs Group, Inc.
C=data.DataReader("C",'yahoo',start,end)
#Berkshire Hathaway Inc.
BLK=data.DataReader("BLK",'yahoo',start,end)
#American Express Company

tickers3=['JPM','GS','C','BLK']
FinancialBanks=pd.concat([JPM, GS, C, BLK],axis=1,keys=tickers3)
FinancialBanks.columns.names=['Stock name','Stock Info']
FinancialBanks.describe()

FinancialBanks
```

```python
NKE=data.DataReader("NKE",'yahoo',start,end)
#NIKE, Inc.
TGT=data.DataReader("TGT",'yahoo',start,end)
#Target Corporation .
WMT=data.DataReader("WMT",'yahoo',start,end)
#Walmart Inc.
TJX=data.DataReader("TJX",'yahoo',start,end)
#The TJX Companies, Inc.
tickers4=['NKE','TGT','WMT','TJX']
Retail=pd.concat([NKE,TGT,WMT,TJX],keys=tickers4,axis=1)
Retail.columns.names=['Stock name','Stock Info']
Retail.describe()

Retail
```

```python
VRTX=data.DataReader("VRTX",'yahoo',start,end)
#Vertex Pharmaceuticals Incorporated 
ABBV=data.DataReader("ABBV",'yahoo',start,end)
#AbbVie Inc.
BSX=data.DataReader("BSX",'yahoo',start,end)
#Bristol-Myers Squibb Company
JNJ=data.DataReader("JNJ",'yahoo',start,end)
#Johnson & Johnson 
tickers5=['VRTX','ABBV','BSX','JNJ']
Healthcare=pd.concat([VRTX,ABBV,BSX,JNJ],keys=tickers5,axis=1)
Healthcare.columns.names=['Stock name','Stock Info']
Healthcare.describe()

Healthcare
```

```python
All=pd.concat([Airlines,Technology,FinancialBanks,Retail,Healthcare],keys=['Airlines','Technology','FinancialBanks','Retail','Healthcare'],axis=1)
```

```python
Airlinereturns=pd.DataFrame()
for tick in tickers1:
    Airlinereturns[tick+' Return'] =Airlines[tick]['Close'].pct_change()
Airlinereturns.head()

#shows the percentage change of closing price for each day
```

```python
Technologyreturns=pd.DataFrame()
for tick in tickers2:
    Technologyreturns[tick+' Return'] =Technology[tick]['Close'].pct_change()
Technologyreturns.head()
```

```python
FinancialBanksreturns=pd.DataFrame()
for tick in tickers3:
    FinancialBanksreturns[tick+' Return'] =FinancialBanks[tick]['Close'].pct_change()
FinancialBanksreturns.head()
```

```python
Retailreturns=pd.DataFrame()
for tick in tickers4:
    Retailreturns[tick+' Return'] =Retail[tick]['Close'].pct_change()
Retailreturns.head()
```

```python
Healthcarereturns=pd.DataFrame()
for tick in tickers5:
    Healthcarereturns[tick+' Return'] =Healthcare[tick]['Close'].pct_change()
Healthcarereturns.head()
```
